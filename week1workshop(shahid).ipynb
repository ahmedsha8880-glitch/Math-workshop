```python
# question- b
def myPow(x, n):
    result = 1
    while n > 0:
        result *= x
        n -= 1
    return result

# Example: myPow(2, 5) → 32
```

```python
# question- c
def SumDigits(n):
    total = 0
    while n > 0:
        total += n % 10   # extract last digit
        n //= 10          # remove last digit
    print(total)

# Example: SumDigits(234) → 9
```

```python
# question- d
def squaredSum(arr):
    total = 0
    for num in arr:
        total += num * num
    return total

# Example: squaredSum([2,3,4]) → 29
```

```python
# question- e
def sumDigits(num):
    total = 0
    while num > 0:
        digit = num % 10
        total += digit * digit   # square each digit before summing
        num //= 10
    return total

# Example: sumDigits(6) → 36 (since 6² = 36)
```
