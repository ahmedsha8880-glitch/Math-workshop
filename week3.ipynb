```python
#Exercise:4
import matplotlib.pyplot as plt
import numpy as np

# Define the range of x values
x = np.linspace(-10, 10, 100)  # from -10 to 10 with 100 points
y = x  # since y = x

# Plot the line
plt.plot(x, y, label="y = x", color="blue")

# Add labels and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph of y = x")
plt.legend()

# Show grid and plot
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # x-axis
plt.axvline(0, color='black', linewidth=0.5)  # y-axis
plt.show()
```

```python
#Exercise:6
import numpy as np
import matplotlib.pyplot as plt

# Generate 20 x values from -10 to 10
x = np.linspace(-10, 10, 20)

# Define the functions
y1 = 5 * x**3 + 2 * x - 1
y2 = -2 * x**3 + x**2 + 100
y3 = 2 * np.pi * x + 20

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='f(x) = 5x³ + 2x - 1', marker='o')
plt.plot(x, y2, label='f(x) = -2x³ + x² + 100', marker='s')
plt.plot(x, y3, label='f(x) = 2πx + 20', marker='^')

# Add labels and legend
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Plot of Three Functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Exercise:7
import numpy as np
import matplotlib.pyplot as plt

# Generate 20 x values
x = np.linspace(-10, 10, 20)

# Define the functions
y1 = 5 * x**3 + 2 * x - 1
y2 = -2 * x**3 + x**2 + 100
y3 = 2 * np.pi * x + 20

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 12))  # 3 rows, 1 column

# Plot each function in its own subplot
axs[0].plot(x, y1, marker='o', color='blue')
axs[0].set_title('f(x) = 5x³ + 2x - 1')
axs[0].grid(True)

axs[1].plot(x, y2, marker='s', color='green')
axs[1].set_title('f(x) = -2x³ + x² + 100')
axs[1].grid(True)

axs[2].plot(x, y3, marker='^', color='red')
axs[2].set_title('f(x) = 2πx + 20')
axs[2].grid(True)

# Add common labels
for ax in axs:
    ax.set_xlabel('x values')
    ax.set_ylabel('y values')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#Exercise:8
import matplotlib.pyplot as plt

# Define the points
x = [1, 2, 6, 8]
y = [3, 8, 1, 10]

# Create subplots: 2 rows, 2 columns
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot in each subplot with different styles
axs[0, 0].plot(x, y, marker='o', color='blue')
axs[0, 0].set_title("Line through points")
axs[0, 0].grid(True)

axs[0, 1].scatter(x, y, color='green')
axs[0, 1].set_title("Scatter plot of points")
axs[0, 1].grid(True)

axs[1, 0].bar(x, y, color='orange')
axs[1, 0].set_title("Bar chart of points")
axs[1, 0].grid(True)

axs[1, 1].step(x, y, color='red')
axs[1, 1].set_title("Step plot of points")
axs[1, 1].grid(True)

# Add common labels
for ax in axs.flat:
    ax.set_xlabel("x-axis")
    ax.set_ylabel("y-axis")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#Exercise:9
import numpy as np
import matplotlib.pyplot as plt

# Define x values from -3π to 3π
x = np.linspace(-3*np.pi, 3*np.pi, 400)  # 400 points for smooth curve
y = np.cos(x)

# Plot the cosine function
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label='y = cos(x)')

# Add labels and title
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Cosine Function in Interval [-3π, 3π]')
plt.legend()
plt.grid(True)

# Show axes lines
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Display the plot
plt.show()
```

```python
#Exercise:10
import numpy as np
import matplotlib.pyplot as plt

# Circle parameters
a, b = 2, 2   # center
r = 4         # radius

# Initialize theta
theta = -np.pi
x_vals = []
y_vals = []

# Use while loop to generate points
while theta <= np.pi:
    x = a + r * np.cos(theta)
    y = b + r * np.sin(theta)
    x_vals.append(x)
    y_vals.append(y)
    theta += 0.1  # step size

# Plot the circle
plt.figure(figsize=(6,6))
plt.plot(x_vals, y_vals, color='blue', label='Circle')
plt.scatter(a, b, color='red', marker='o', label='Center (2,2)')
plt.gca().set_aspect('equal', adjustable='box')

# Add labels and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Circle with center (2,2) and radius 4")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Exercise:11
import numpy as np
import matplotlib.pyplot as plt

# Parameters for circle
a, b = 2, 2   # center
r = 4         # radius

# Parameters for ellipse
A, B = 6, 3   # semi-major and semi-minor axes

# Generate theta values
theta = np.linspace(0, 2*np.pi, 400)

# Circle coordinates
x_circle = a + r * np.cos(theta)
y_circle = b + r * np.sin(theta)

# Ellipse coordinates
x_ellipse = a + A * np.cos(theta)
y_ellipse = b + B * np.sin(theta)

# Plot both
plt.figure(figsize=(7,7))
plt.plot(x_circle, y_circle, label="Circle (center (2,2), r=4)", color="blue")
plt.plot(x_ellipse, y_ellipse, label="Ellipse (center (2,2), A=6, B=3)", color="red")

# Mark the center
plt.scatter(a, b, color="black", marker="o", label="Center (2,2)")

# Equal aspect ratio for proper shape
plt.gca().set_aspect('equal', adjustable='box')

# Labels and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Circle and Ellipse on Same Axes")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Exercise:12
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p, q, a = 2, 3, 1

# Generate x values
x = np.linspace(-10, 10, 400)

# Parabola equation
y = 4*a*(x - q)**2 + p

# Plot the parabola
plt.figure(figsize=(8,6))
plt.plot(x, y, color='blue', label=f'Parabola: (y - {p}) = 4*{a}(x - {q})^2')

# Mark the vertex
plt.scatter(q, p, color='red', marker='o', label=f'Vertex ({p},{q})')

# Labels and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Parabola with Vertex at (p,q)")
plt.legend()
plt.grid(True)
plt.show()
```
