```python
import numpy as np

def dot_product(vec1, vec2):
    # flatten the input vectors
    vec1 = vec1.flatten()
    vec2 = vec2.flatten()
    return np.sum(np.multiply(vec1, vec2))

def matrix_vector_product(mat, vec):
    m, n = mat.shape
    p, q = vec.shape

    assert n == p, "The matrix and vector are not compatible for multiplication!"

    # create a ndarray to store the result
    result = np.zeros((m, q), dtype=float)

    # Loop to perform Matrix-Vector product
    for i in range(m):
        result[i] = dot_product(mat[i, :], vec)

    return result


if __name__ == "__main__":

    # Part 2: Problem 2
    mat1 = np.full((3,3), 3, dtype=int)   # 3×3 matrix filled with 3
    vec1 = np.full((3,1), 2, dtype=int)   # 3×1 vector filled with 2

    print("Matrix = ")
    print(mat1)
    print("Vector = ")
    print(vec1)

    # Call the function
    result1 = matrix_vector_product(mat1, vec1)
    print("Result (Manual) = ")
    print(result1)

    # Check with numpy built-in function
    result2 = mat1 @ vec1
    print("Result (NumPy) = ")
    print(result2)
```

```python
import numpy as np

def dot_product(vec1, vec2):
    vec1 = vec1.flatten()
    vec2 = vec2.flatten()
    return np.sum(np.multiply(vec1, vec2))


def matrix_matrix_product(A, B):
    m, n = A.shape
    p, q = B.shape

    assert n == p, "Matrices are not compatible for multiplication!"

    # Result matrix of shape (m x q)
    result = np.zeros((m, q), dtype=float)

    # Compute matrix-matrix product
    for i in range(m):
        for j in range(q):
            result[i, j] = dot_product(A[i, :], B[:, j])

    return result


if __name__ == "__main__":

    # Example matrices
    A = np.array([[1, 2, 3],
                  [4, 5, 6]])

    B = np.array([[7, 8],
                  [9, 10],
                  [11, 12]])

    print("Matrix A =")
    print(A)

    print("Matrix B =")
    print(B)

    # Custom result
    custom_result = matrix_matrix_product(A, B)
    print("Custom Matrix-Matrix Product =")
    print(custom_result)

    # NumPy built-in result
    numpy_result = A @ B
    print("NumPy dot() Result =")
    print(numpy_result)
```

```python
import numpy as np

def is_symmetric(matrix):
    # A matrix is symmetric if it is equal to its transpose
    return np.array_equal(matrix, matrix.T)


if __name__ == "__main__":

    # Example matrices
    A = np.array([[1, 2, 3],
                  [2, 5, 6],
                  [3, 6, 9]])

    B = np.array([[1, 0, 2],
                  [0, 5, 7],
                  [3, 7, 9]])

    print("Matrix A =")
    print(A)
    print("Is A symmetric? ->", is_symmetric(A))

    print("\nMatrix B =")
    print(B)
    print("Is B symmetric? ->", is_symmetric(B))
```
